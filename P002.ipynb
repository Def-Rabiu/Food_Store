```python
import tkinter
from tkinter import *
import tkinter.messagebox
import customtkinter
import time
from PIL import Image, ImageTk

customtkinter.set_appearance_mode("Dark")  # Modes: "System" (standard), "Dark", "Light"
customtkinter.set_default_color_theme("green")  # Themes: "blue" (standard), "green", "dark-blue"

class App(customtkinter.CTk):

    WIDTH = 1140
    HEIGHT = 740

    def __init__(self):
        super().__init__()

        self.Undo = []
        self.Redo = []
        self.food_order = {}

        
        self.count = 0

        self.title("Shopping cart")
        self.geometry(f"{App.WIDTH}x{App.HEIGHT}")
        self.protocol("WM_DELETE_WINDOW", self.on_closing)  # call .on_closing() when app gets closed

        #create columns $ rows
        self.grid_columnconfigure(1, weight=1)
        self.grid_rowconfigure(0, weight=1)

        #get left frame
        self.frame_left = customtkinter.CTkFrame(master=self, width=350, corner_radius=0)
        self.frame_left.grid(row=0, column=0, sticky="nswe")
        self.frame_left.configure(fg_color="#333")
        #Get right frame
        self.frame_right = customtkinter.CTkFrame(master=self)
        self.frame_right.grid(row=0, column=1, sticky="nswe", padx=20, pady=20)
        # # Adding an image
        self.cart_image = ImageTk.PhotoImage(Image.open("shoping_cart.png").resize((20,20), Image.Resampling.LANCZOS))
        self.button_head = customtkinter.CTkButton(master=self.frame_right, image=self.cart_image,
                                                    text="Shopping Cart", text_color_disabled="#222",            
                                                    width=220, height=60, compound="top", state=tkinter.DISABLED)
        self.button_head.grid(row=0, column=0, columnspan=5, pady=20, padx=20, sticky="we") 


        
        # TAke order input
        self.entry1 = customtkinter.CTkEntry(master=self.frame_right, placeholder_text="Type in your order")
        self.entry1.grid(row=2, column=0, columnspan=3, pady=10, padx=10, sticky="we")
        # Add Button
        self.button_add = customtkinter.CTkButton(master=self.frame_right,
                                                text="Add item", #button label text
                                                border_width=2,  # <- custom border_width
                                                fg_color=None,  # <- no fg_color
                                                command=self.push)
        self.button_add.grid(row=3, column=0, columnspan=2, pady=20, padx=20, sticky="we")
        # Quantity input
        self.entry2 = customtkinter.CTkEntry(master=self.frame_right, width=100, placeholder_text="Order Quantity")
        self.entry2.grid(row=2, column=3, columnspan=2, pady=10, padx=10, sticky="we")
        # Undo Button
        self.button_undo = customtkinter.CTkButton(master=self.frame_right,
                                                text="Undo", hover_color="#d35B58",
                                                border_width=2,  # <- custom border_width
                                                fg_color=None,  # <- no fg_color
                                                command=self.undo)
        self.button_undo.grid(row=3, column=2, columnspan=1, pady=20, padx=20, sticky="we")
        #  Redo Button 
        self.button_redo = customtkinter.CTkButton(master=self.frame_right,
                                                text="Redo", #button label text
                                                border_width=2,  # <- custom border_width
                                                fg_color=None,  # <- no fg_color
                                                command=self.redo)
        self.button_redo.grid(row=3, column=3, columnspan=1, pady=20, padx=20, sticky="we")

        # Available items Popup
        self.pop_button = customtkinter.CTkButton(self.frame_right, text="Click To See Available Items", command=self.create_toplevel)
        self.pop_button.grid(row=4, column=0, columnspan=4, pady=20, padx=20, sticky="we")

        self.checkout = customtkinter.CTkButton(master=self.frame_right,
                                                text="Checkout", #button label text
                                                border_width=2,  # <- custom border_width
                                                fg_color=None,  # <- no fg_color
                                                command=self.checkout)
        self.checkout.grid(row=6, column=2, columnspan=1, pady=20, padx=20, sticky="we") 

    def push(self):
        inp = self.entry1.get().title()
        inp2 = self.entry2.get()
        if inp2.isdigit():
          print(inp2)
          inp2 = int(inp2)
        else:
          inp2 = 1
        self.Undo.append((inp, inp2))
        self.count +=1
        print(self.Undo)
        for e in self.Undo:
          text_var = tkinter.StringVar(value=f"ITEM -> {e[0]}   ||   QTY -> {e[1]}")
          self.label = customtkinter.CTkLabel(master=self.frame_left,
                                textvariable=text_var,
                                width=300,
                                fg_color="black",
                                relief="solid",
                                corner_radius=9)
          self.label.grid(row=self.count, column=0, pady=20, padx=20, sticky="we")
    
    def create_toplevel(self):
      self.popup = customtkinter.CTkToplevel(self)
      self.popup.geometry("400x250")
      self.popup.title("WE HAVE:")

      # create label on CTkToplevel window
      label = customtkinter.CTkLabel(self.popup, text="Beans -> $200")
      label.pack(side="top", fill="both", padx=10, pady=10)

      label2 = customtkinter.CTkLabel(self.popup, text="Rice -> $100")
      label2.pack(side="top", fill="both", padx=10, pady=10)

      label3 = customtkinter.CTkLabel(self.popup, text="Plantain -> $200")
      label3.pack(side="top", fill="both", padx=10, pady=10)

      # label4 = customtkinter.CTkLabel(self.popup, text="Egg -> $100")
      # label4.pack(side="top", fill="both", padx=10, pady=10)
      
      label5 = customtkinter.CTkLabel(self.popup, text="Turkey -> $500")
      label5.pack(side="top", fill="both", padx=10, pady=10)

      label6 = customtkinter.CTkLabel(self.popup, text="Bread -> $150")
      label6.pack(side="top", fill="both", padx=10, pady=10)

      label7 = customtkinter.CTkLabel(self.popup, text="Bread -> $150")
      label7.pack(side="top", fill="both", padx=10, pady=10)

    def undo(self):
      try:
          x = self.Undo[-1]
          popped = self.Undo.pop()
          self.Redo.append(x)
          print("undo", self.Undo)
          text_var = tkinter.StringVar(value="")
          self.label = customtkinter.CTkLabel(master=self.frame_left,
                                textvariable=text_var,
                                width=300,
                                corner_radius=8)
          self.label.grid(row=self.count, column=0, pady=20, padx=20, sticky="we")
          self.count -=1
          # tk.Label(window, anchor=tk.W, text=Undo.pop()).grid(row=count-1, column=1)
      except:
        print("its overrrrr")
        text_var = tkinter.StringVar(value="")
        self.label = customtkinter.CTkLabel(master=self.frame_left,
                                textvariable=text_var,
                                width=300,
                                corner_radius=8)
        self.label.grid(row=self.count, column=0, pady=20, padx=20, sticky="we")
          

    def redo(self):
      try:
          x = self.Redo[-1]
          self.Redo.pop()
          self.Undo.append(x)
          print("redo", self.Undo)
          self.count +=1
          for e in self.Undo:
            text_var = tkinter.StringVar(value=f"ITEM -> {e[0]}   ||   QTY -> {e[1]}")
            self.label = customtkinter.CTkLabel(master=self.frame_left,
                                  textvariable=text_var,
                                  width=350,
                                  fg_color="black",
                                  relief="solid",
                                  corner_radius=8)
            self.label.grid(row=self.count, column=0, pady=20, padx=20, sticky="we")
          
          # tk.Label(window, anchor=tk.W, text=)).grid(row=count, column=1)
      except:
          print("nothing to print")

    # tthis function handles the available orders file
    # and stores it in a dictionary: STORE
    # it returns the text file saved in a dictionary
    # TIME COMPLEXITY: O(n
    def open_file(self):
      store = {}
      try:
          with open("food.txt") as orders:
              for lines in orders:
                  order, price = lines.strip().split(":")
                  store[order] = int(price)
      except FileNotFoundError:
          print("order not found chief!")
      else:
        return store

    # this function takes care of the user request 
    # and it returns a ditionary: FOOD_ORDER
    # containing the requests
    # TIME COMPLEXITY: O(n)
    def order_function(self):
      food_order ={}
      if self.Undo is not None:
          for ordr in self.Undo:
              food_order[ordr[0]] = int(ordr[1])
      else:
          print("enter a valid order")
      return food_order

    # this function returns a list of the user orders
    # for the purpose of calculating the quantity and prices
    # TIME COMPLEXITY: O(1)
    def list_of_orders(self):
      list = []
      if not self.order_function():
          return
      else:
          for k,v in self.order_function().items():
              a = [k] * v
              for items in a:
                  list.append(items)
      return list
    
    # comparing what we have in our available list of foods 
    # and what the user has ordered
    # TIME COMPLEXITY: O(n)
    def store_function(self):
      sorted_list = []
      unavailable=  set()
      for food_ordered in self.list_of_orders():
          if food_ordered in self.open_file():
              sorted_list.append(food_ordered)
          else:
              unavailable.add(food_ordered)
      return unavailable

    # # this function sums up the amount for the quantities of food we have order!
    # # this is end product
    # TIME COMPLEXITY: O(n^2)
    def sum_order(self):
      dict2 = {}
      for k,v in self.open_file().items():
          for food in self.list_of_orders():
              if food == k:
                  dict2[food] = dict2.get(food, 0) + v
      return dict2

    def checkout(self):
      order_sum = self.sum_order()
      check_unavailable = self.store_function()
      total = 0
      row = 7
      today = time.strftime('%b|%d|%R')
      #available items
      for k,v in order_sum.items():
        text_var = tkinter.StringVar(value=f"{k}     ->  sub-total: ${v} :  {today}")
        self.label_aval = customtkinter.CTkLabel(master=self.frame_right,
                                  textvariable=text_var,
                                  fg_color="black",
                                  relief="solid",
                                  corner_radius=8)
        self.label_aval.grid(row=row, column=0, pady=10, padx=10, sticky="we")
        total += v
        row += 1
      text_var = tkinter.StringVar(value=f"    Total:     ${total}")
      self.label_aval = customtkinter.CTkLabel(master=self.frame_right,textvariable=text_var,fg_color="black",relief="solid",corner_radius=8)
      self.label_aval.grid(row=row, column=0, pady=10, padx=10, sticky="we")
        #Unavailable items
      row2 = 7
      for item in check_unavailable:
        text_var = tkinter.StringVar(value=f"{item} is not available")
        self.label_unavailable = customtkinter.CTkLabel(master=self.frame_right,
                                textvariable=text_var,
                                fg_color="black",
                                relief="solid",
                                corner_radius=8)
        self.label_unavailable.grid(row=row2, column=3, pady=10, padx=10, sticky="we")
        row2 += 1

    def on_closing(self, event=0):
        self.destroy()

if __name__ == "__main__":
    app = App()
    app.mainloop()

```
